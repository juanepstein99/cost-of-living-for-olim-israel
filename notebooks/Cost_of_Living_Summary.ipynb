# Cost of Living Summary

This notebook extends the rent analysis into a broader cost-of-living analysis.

The main goal is to compare affordability across four Israeli cities:

- Tel Aviv
- Ramat Gan
- Haifa
- Jerusalem

The main KPI of this project is:

**Rent-to-Salary Ratio**

This shows what percentage of an average monthly salary is spent on rent.

# Methodology

## Data Sources

### Rent
Source: Madlan rental listings.

### Salary
Source:
Average gross monthly salary data was obtained from the National Insurance Institute (Bituach Leumi) using the most recent city-level figures available (2024).

Gross salary was used instead of estimated net salary to avoid introducing additional assumptions related to taxes, tax credits, pension contributions, and personal circumstances.

Using the same gross salary basis across all cities ensures a consistent comparison of affordability.
### Transportation
Source: Rav-Kav Online monthly public transportation pass.

### Grocery Basket
Source: Numbeo grocery prices.

The food basket is used as a reference price comparison across cities.
It does not represent a full monthly grocery budget.

### Arnona
Source: Municipal tax ordinances published by the Municipality of Tel Aviv-Yafo, Municipality of Ramat Gan, Municipality of Haifa, and Municipality of Jerusalem.

## Key Assumptions

- Official 2024 gross salary data was used for all cities.
- The same reference food basket was applied across all cities to compare grocery price levels.
- The reference food basket is not included in the total monthly cost calculation because it does not represent a full monthly grocery budget.
- Transportation costs were based on a standard monthly public transportation pass.
- Monthly Arnona was estimated using the municipal residential tax rate (₪ per sqm per year) and the median apartment size observed in each city's rental dataset.
- Affordability was measured as the percentage of average gross salary required to pay rent.

## Important Note on Rent Metrics

Rental prices are not normally distributed and include a small number of luxury properties, particularly in Jerusalem.

These listings were retained because they represent real market observations rather than data errors. As a result, average rent may be higher than median rent in some cities.

For this reason, both average and median rent values are considered during the analysis. While average rent reflects the overall market, median rent provides a more representative estimate of the typical rental experience for young immigrants.

This distinction is important when interpreting affordability metrics and comparing cities.

```python
# Import required libraries
import pandas as pd
import numpy as np
```

## 1. Load City Rent Summary

This dataset was created in the Master Rent Dataset notebook.

It contains city-level rent KPIs after cleaning and outlier filtering.

```python
# Load city rent summary
rent_summary = pd.read_csv("city_rent_summary.csv")

# Preview dataset
rent_summary
```

## 2. Create Cost of Living Input Table

This table will contain the additional cost-of-living indicators.

The values will be filled manually from consistent sources.

Columns:

- reference_food_basket_cost: reference grocery basket used for price comparison
- monthly_transport: monthly public transportation cost
- average_salary: average monthly salary
- arnona_rate_per_sqm_yearly: yearly Arnona rate per square meter

```python
# Create manual cost of living table
city_cost = pd.DataFrame({
    "City": ["Tel Aviv", "Ramat Gan", "Haifa", "Jerusalem"],

    # Reference grocery basket for price comparison
    "reference_food_basket_cost": [np.nan, np.nan, np.nan, np.nan],

    # Monthly public transportation cost
    "monthly_transport": [np.nan, np.nan, np.nan, np.nan],

    # Average monthly salary
    "average_salary": [np.nan, np.nan, np.nan, np.nan],

    # Yearly Arnona rate per square meter
    "arnona_rate_per_sqm_yearly": [np.nan, np.nan, np.nan, np.nan]
})

city_cost
```

## 3. Data Sources and Methodology

The cost-of-living data will be collected using consistent sources:

- Salary: CBS / official salary data when available
- Transportation: public transportation monthly pass estimate
- Reference Food Basket: one consistent basket estimate source
- Arnona: official municipality residential rates

### Reference Food Basket

The food basket was created using a fixed list of common grocery items from Numbeo:

milk, bread, eggs, rice, chicken, tomatoes, potatoes, apples, bananas, and water.

This basket is used as a reference price comparison across cities.
It is not treated as a full monthly grocery budget and is therefore not included in the total monthly cost or disposable income calculations.

## 4. Fill Cost of Living Values

In this section, I manually enter the collected values for each city.

```python
# Tel Aviv values
city_cost.loc[city_cost["City"] == "Tel Aviv", "reference_food_basket_cost"] = 145.45
city_cost.loc[city_cost["City"] == "Tel Aviv", "monthly_transport"] = 323
city_cost.loc[city_cost["City"] == "Tel Aviv", "average_salary"] = 21283
city_cost.loc[city_cost["City"] == "Tel Aviv", "arnona_rate_per_sqm_yearly"] = 90

city_cost
```

```python
# Ramat Gan values
city_cost.loc[city_cost["City"] == "Ramat Gan", "reference_food_basket_cost"] = 121.22
city_cost.loc[city_cost["City"] == "Ramat Gan", "monthly_transport"] = 323
city_cost.loc[city_cost["City"] == "Ramat Gan", "average_salary"] = 19282
city_cost.loc[city_cost["City"] == "Ramat Gan", "arnona_rate_per_sqm_yearly"] = 81

city_cost
```

```python
# Haifa values
city_cost.loc[city_cost["City"] == "Haifa", "reference_food_basket_cost"] = 125.68
city_cost.loc[city_cost["City"] == "Haifa", "monthly_transport"] = 323
city_cost.loc[city_cost["City"] == "Haifa", "average_salary"] = 15854
city_cost.loc[city_cost["City"] == "Haifa", "arnona_rate_per_sqm_yearly"] = 69

city_cost
```

```python
# Jerusalem values
city_cost.loc[city_cost["City"] == "Jerusalem", "reference_food_basket_cost"] = 127.29
city_cost.loc[city_cost["City"] == "Jerusalem", "monthly_transport"] = 323
city_cost.loc[city_cost["City"] == "Jerusalem", "average_salary"] = 11480
city_cost.loc[city_cost["City"] == "Jerusalem", "arnona_rate_per_sqm_yearly"] = 65
city_cost
```

## 5. Merge Rent Data With Cost of Living Data

Now I combine the rent summary with the additional cost indicators.

```python
# Merge rent data with cost of living indicators
cost_of_living = rent_summary.merge(
    city_cost,
    on="City",
    how="left"
)

cost_of_living
```

## 6. Add Estimated Apartment Size

Median apartment size was estimated using the cleaned rental datasets for each city.

These values are later used to estimate monthly Arnona costs based on municipal residential tax rates.

```python
# Estimated average apartment size by city
avg_size = pd.DataFrame({
    "City": ["Tel Aviv", "Ramat Gan", "Haifa", "Jerusalem"],
    "average_size_sqm": [65, 65, 70, 65]
})

avg_size
```

```python
# Merge estimated apartment size
cost_of_living = cost_of_living.merge(
    avg_size,
    on="City",
    how="left"
)

cost_of_living
```

## 7. Estimate Monthly Arnona

Arnona is a municipal tax that varies by city, property type, area and apartment size.

For this project, I use a consistent estimation method:

estimated_monthly_arnona = average_size_sqm × arnona_rate_per_sqm_yearly / 12

```python
# Calculate estimated monthly Arnona
cost_of_living["estimated_monthly_arnona"] = (
    cost_of_living["average_size_sqm"] *
    cost_of_living["arnona_rate_per_sqm_yearly"]
) / 12

# Round result
cost_of_living["estimated_monthly_arnona"] = (
    cost_of_living["estimated_monthly_arnona"].round(2)
)

cost_of_living
```

## 8. Main KPI: Housing Affordability

The main KPI of this project is the Rent-to-Salary Ratio.

rent_to_salary_ratio = median_rent / average gross salary

This shows what percentage of an average monthly salary is required to pay rent.

Lower values indicate better affordability.

```python
# Calculate rent-to-salary ratio
cost_of_living["rent_to_salary_ratio"] = (
    cost_of_living["median_rent"] /
    cost_of_living["average_salary"]
)

# Convert ratio to percentage
cost_of_living["rent_to_salary_percent"] = (
    cost_of_living["rent_to_salary_ratio"] * 100
)

cost_of_living[[
    "City",
    "average_rent",
    "average_salary",
    "rent_to_salary_ratio",
    "rent_to_salary_percent"
]]
```

## 9. Total Monthly Cost and Disposable Income

In this section, I estimate the total monthly cost of living.

total_monthly_cost = rent + transport + estimated Arnona

disposable_income = average gross salary - total monthly cost

The reference food basket is excluded from total monthly cost because it is used only as a price comparison indicator, not as a full monthly grocery budget.

```python
# Calculate total estimated monthly cost
cost_of_living["total_monthly_cost"] = (
    cost_of_living["median_rent"] +
    cost_of_living["monthly_transport"] +
    cost_of_living["estimated_monthly_arnona"]
)

# Calculate pre-tax disposable income after rent, transportation, and Arnona
cost_of_living["disposable_income"] = (
    cost_of_living["average_salary"] -
    cost_of_living["total_monthly_cost"]
)

# Round final KPI columns
cols_to_round = [
    "estimated_monthly_arnona",
    "rent_to_salary_ratio",
    "rent_to_salary_percent",
    "total_monthly_cost",
    "disposable_income"
]

cost_of_living[cols_to_round] = cost_of_living[cols_to_round].round(2)

cost_of_living
```

## 10. Final Cost of Living Dataset

This final table will be exported for Tableau.

It includes rent, basic cost indicators and affordability KPIs.

```python
# Select final columns
city_cost_summary = cost_of_living[[
    "City",
    "average_rent",
    "median_rent",
    "average_price_per_room",
    "average_price_per_sqm",
    "reference_food_basket_cost",
    "monthly_transport",
    "average_salary",
    "average_size_sqm",
    "arnona_rate_per_sqm_yearly",
    "estimated_monthly_arnona",
    "rent_to_salary_ratio",
    "rent_to_salary_percent",
    "total_monthly_cost",
    "disposable_income"
]]

city_cost_summary
```

## 11. Quick Ranking Checks

Before exporting, I check the main affordability rankings.

```python
# Sort cities by rent-to-salary percentage
city_cost_summary.sort_values(
    "rent_to_salary_percent",
    ascending=True
)[[
    "City",
    "average_rent",
    "average_salary",
    "rent_to_salary_percent"
]]
```

```python
# Sort cities by disposable income
city_cost_summary.sort_values(
    "disposable_income",
    ascending=False
)[[
    "City",
    "total_monthly_cost",
    "average_salary",
    "disposable_income"
]]
```

## 12. Export Final Dataset

The final CSV will be used in Tableau for the cost-of-living and affordability dashboards.

```python
# Export final dataset
city_cost_summary.to_csv("city_cost_summary.csv", index=False)

print("city_cost_summary.csv exported successfully.")
```
